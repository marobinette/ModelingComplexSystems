### Euler's method

```python
import matplotlib.pyplot as plt
```

```python
"""
    Implementation of Euler's method

    The below implementation is of the example in ch 6.4 of the textbook. It can be modified for other models by changing the initial conditions and dxdt.
"""

def euler():
    # Initialize
    h = 0.01 # Step size
    T = 50 # Time to estimate
    x = 0.1 # Starting x value
    r, K = 0.2, 1 # Time-invariant parameters
    points = [x] # Stores x values

    # Run Euler method for T/h time steps (floor div in case T/h is not an int)
    for i in range(int(T//h)):
        # Get next point
        points.append(points[-1] + h*euler_dxdt(points[-1],r,K))
    return points, h

def euler_dxdt(x,r,K):
    return r*x*(1-(x/K))

points, h = euler()
print(points)
timesteps = [i*h for i in range(len(points))]
plt.plot(timesteps, points)
plt.show()
```

### Heun's method

```python
"""
    Implementation of Heun's method

    The below implementation is of the example in ch 6.4 of the textbook. It can be modified for other models by changing the initial conditions and dxdt.
"""

def heun():
    # Initialize
    h = 0.01 # Step size
    T = 50 # Time to estimate
    x = 0.1 # Starting x value
    r, K = 0.2, 1 # Time-invariant parameters
    points = [x] # Stores x values
    
    # Run Heun's method for T/h time steps (floor div in case T/h is not an int)
    for i in range(int(T//h)):
        # First step: Euler step
        euler = points[-1] + h*heun_dxdt(points[-1],r,K)
        # Second step: Heun prediction
        points.append(points[-1] + (h/2)*(heun_dxdt(points[-1],r,K) + heun_dxdt(euler,r,K)))
    return points, h

def heun_dxdt(x,r,K):
    return r*x*(1-(x/K))

points, h = heun()
print(points)
timesteps = [i*h for i in range(len(points))]
plt.plot(timesteps, points)
plt.show()
```
