```python
from pylab import *

class Oscillation:
    def __init__(self, x, y, x_result, y_result):
        self.x = x
        self.y = y
        self.x_result = x_result
        self.y_result = y_result
        self.x_result.append(x)
        self.y_result.append(y)
        
    def observe(self):
        self.x_result.append(self.x)
        self.y_result.append(self.y)    

    def update(self):
        self.next_x = 0.5 * self.x + self.y
        self.next_y = -0.5 * self.x + self.y
        self.x = self.next_x
        self.y = self.next_y
```

```python
oscillation = Oscillation(x=1, y=1, x_result=[], y_result=[])
for t in range(30):
    oscillation.update()
    oscillation.observe()

plot(oscillation.x_result, 'b-')
plot(oscillation.y_result, 'g--')
show()
```
