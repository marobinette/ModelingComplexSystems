```python
class ExponentialGrowth:
    def __init__(self, x, result, a, t, time_steps):
        self.x = x
        self.result = result
        self.result.append(x)
        self.a = a
        self.t = t
        self.time_steps = time_steps
        self.time_steps.append(t)
        
    def observe(self):
        self.result.append(self.x)
        self.time_steps.append(self.t)      

    def update(self):
        self.x = self.a * self.x
        self.t += 0.1
```

```python
exponential_growth = ExponentialGrowth(x=1, result=[], a=1.1, t=0, time_steps=[])

while exponential_growth.t < 3:
    exponential_growth.update()
    exponential_growth.observe()

print(exponential_growth.result)

```
